### Estandarizacion

```python
# --- Celda para Estandarizar y Unificar Datasets ---

logging.info("Iniciando proceso de estandarización y unificación de datasets de visitantes.")

# Diccionario de mapeo universal. Contiene todas las variaciones encontradas.
column_mapping = {
    'Encuesta': 'id_encuesta',
    'Fecha': 'fecha',
    'Mes': 'mes',
    'País de procedencia': 'pais_procedencia',
    'Pais de Procedencia': 'pais_procedencia', # Variación de 2018
    'Pais de procedencia': 'pais_procedencia', # Variación de 2019
    'Cuidad': 'ciudad_procedencia',
    'Ciudad': 'ciudad_procedencia', # Corrección de 2019
    'Procedencia': 'tipo_procedencia',
    'Zona procedendia- Visitantes nacionales': 'zona_procedencia_nacional',
    'Zona de procedencia-Visitantes  nacionales': 'zona_procedencia_nacional', # Variación de 2018/19
    'Zona procedendia- Visitantes internacionales': 'zona_procedencia_internacional',
    'Zona de procedencia- Visitantes internacionales': 'zona_procedencia_internacional', # Variación de 2018
    'Sexo': 'genero',
    'Edad': 'edad',
    'Edad rangos': 'rango_edad',
    'Motivo Viaje': 'motivo_viaje_principal',
    'Otro motivo de viaje o consulta PIT-Cual?': 'motivo_viaje_secundario',
    'Idioma ': 'idioma_principal', # Con espacio al final
    'Idioma': 'idioma_principal',
    'Otro idioma-Cual?': 'idioma_secundario',
    'Otro idioma Cual?': 'idioma_secundario', # Variación de 2018/19
    'Alojamiento': 'tipo_alojamiento_principal',
    'Otro alojamiento-Cual?': 'tipo_alojamiento_secundario',
    'Otro alojamiento Cual?': 'tipo_alojamiento_secundario', # Variación de 2018/19
    'Tipo de transporte': 'tipo_transporte',
    'Tiempo de estancia': 'tiempo_estancia_dias',
    'Rango de Tiempo de estancia': 'rango_estancia',
    'Rango de Tiempo': 'rango_estancia' # Variación de 2019
}

# Lista final de columnas que queremos en nuestro dataset unificado
final_columns = [
    'id_encuesta', 'fecha', 'mes', 'pais_procedencia', 'ciudad_procedencia', 
    'tipo_procedencia', 'zona_procedencia_nacional', 'zona_procedencia_internacional',
    'genero', 'edad', 'rango_edad', 'motivo_viaje_principal', 'motivo_viaje_secundario',
    'idioma_principal', 'idioma_secundario', 'tipo_alojamiento_principal', 
    'tipo_alojamiento_secundario', 'tipo_transporte', 'tiempo_estancia_dias', 'rango_estancia'
]

# Lista de archivos a procesar
data_folder = '../datos_brutos/Local_(Cali)/'
visitor_files = [
    '004_Local_Cali_Visitantes_2015.csv',
    '005_Local_Cali_Visitantes_2016.csv',
    '006_Local_Cali_Visitantes_2017.csv',
    '007_Local_Cali_Visitantes_2018.csv',
    '008_Local_Cali_Visitantes_2019.csv'
]

# Lista para almacenar los DataFrames limpios
dfs_limpios = []

for file in visitor_files:
    file_path = os.path.join(data_folder, file)
    try:
        # --- LÓGICA DE CARGA MEJORADA ---
        try:
            # Intenta leer el archivo de la forma estándar (UTF-8, separador por coma)
            df = pd.read_csv(file_path)
            logging.info(f"Archivo '{file}' cargado con configuración estándar.")
        except (UnicodeDecodeError, pd.errors.ParserError):
            # Si falla el intento estándar, reintenta con la configuración para archivos especiales
            logging.warning(f"Lectura estándar falló para '{file}'. Reintentando con encoding 'latin1' y separador ';'.")
            df = pd.read_csv(file_path, encoding='latin1', sep=';')
        
        # --- EL RESTO DEL CÓDIGO ES IGUAL ---
        
        # Renombrar columnas usando el diccionario de mapeo
        df.rename(columns=column_mapping, inplace=True)
        
        # Añadir columna del año para mantener la referencia
        df['anio'] = file.split('_')[-1].replace('.csv', '')

        # Seleccionar solo las columnas que existen en el DataFrame actual y están en la lista final
        columnas_existentes = [col for col in final_columns if col in df.columns]
        df_temp = df[columnas_existentes + ['anio']] # Añadimos el año
        
        dfs_limpios.append(df_temp)
        logging.info(f"Archivo '{file}' procesado y estandarizado.")

    except FileNotFoundError:
        logging.error(f"Archivo no encontrado: {file_path}")
    except Exception as e:
        logging.error(f"Error fatal procesando el archivo {file}: {e}")

# Unificar todos los DataFrames en uno solo
if dfs_limpios:
    df_visitantes_unificado = pd.concat(dfs_limpios, ignore_index=True, sort=False)
    logging.info("Todos los datasets de visitantes han sido unificados.")
    
    # Mostrar la información del nuevo DataFrame unificado
    print("--- Información del DataFrame Unificado ---")
    df_visitantes_unificado.info()
    
    print("\n--- Primeras filas del DataFrame Unificado ---")
    print(df_visitantes_unificado.head(50))

else:
    logging.warning("No se procesaron datasets. El DataFrame unificado está vacío.")
```

```python
# --- Celda para Limpieza y Conversión de Tipos de Datos ---

logging.info("Iniciando la conversión de tipos de datos del DataFrame unificado.")

# Copiamos el DataFrame para mantener un respaldo del original
df_limpio = df_visitantes_unificado.copy()

# 1. Conversión de Fechas
# Convertimos la columna 'fecha' a datetime. Los errores se convertirán en NaT (Not a Time).
df_limpio['fecha'] = pd.to_datetime(df_limpio['fecha'], errors='coerce')
logging.info("Columna 'fecha' convertida a datetime.")

# 2. Conversión de Columnas Numéricas
# Convertimos 'edad' y 'tiempo_estancia_dias' a numérico. 
# 'coerce' transformará cualquier valor no numérico (ej. texto) en NaN.
df_limpio['edad'] = pd.to_numeric(df_limpio['edad'], errors='coerce')
df_limpio['tiempo_estancia_dias'] = pd.to_numeric(df_limpio['tiempo_estancia_dias'], errors='coerce')

# Ahora, convertimos a enteros que permiten nulos (Int64)
df_limpio['id_encuesta'] = df_limpio['id_encuesta'].astype('Int64')
df_limpio['edad'] = df_limpio['edad'].astype('Int64')
df_limpio['tiempo_estancia_dias'] = df_limpio['tiempo_estancia_dias'].astype('Int64')
df_limpio['anio'] = df_limpio['anio'].astype('int16')
logging.info("Columnas numéricas ('id_encuesta', 'edad', 'tiempo_estancia_dias', 'anio') convertidas.")

# 3. Conversión de Categóricas Ordenadas
# Definimos el orden lógico para 'rango_edad'
orden_edad = ['0 - 20 Años', '21 - 45 años', 'Más de 45 años']
df_limpio['rango_edad'] = pd.Categorical(df_limpio['rango_edad'], categories=orden_edad, ordered=True)

# Definimos el orden lógico para 'rango_estancia'
orden_estancia = ['1 Noche', '2-3 Noches', '4-6 Noches', '7-10 Noches', 'Más de 10 noches', 'Reside']
df_limpio['rango_estancia'] = pd.Categorical(df_limpio['rango_estancia'], categories=orden_estancia, ordered=True)
logging.info("Columnas categóricas ordenadas ('rango_edad', 'rango_estancia') convertidas.")

# 4. Conversión de Categóricas Simples
# Hacemos una lista de todas las demás columnas de tipo 'object' que deben ser 'category'
columnas_a_categoria = [
    'mes', 'pais_procedencia', 'ciudad_procedencia', 'tipo_procedencia', 
    'zona_procedencia_nacional', 'zona_procedencia_internacional', 'genero', 
    'motivo_viaje_principal', 'motivo_viaje_secundario', 'idioma_principal', 
    'idioma_secundario', 'tipo_alojamiento_principal', 'tipo_alojamiento_secundario', 
    'tipo_transporte'
]

for col in columnas_a_categoria:
    if col in df_limpio.columns:
        df_limpio[col] = df_limpio[col].astype('category')
logging.info("Columnas categóricas simples convertidas.")

# --- Verificación Final ---
# Mostramos la información del DataFrame para confirmar los nuevos tipos de datos
print("\n--- Información del DataFrame con Tipos de Datos Corregidos ---")
df_limpio.info()

logging.info("Proceso de conversión de tipos de datos finalizado.")
```

```python
# --- 4. Eliminar Filas Duplicadas ---
logging.info("Buscando y eliminando filas duplicadas.")

duplicados_antes = df_limpio.duplicated().sum()
logging.info(f"Se encontraron {duplicados_antes} filas duplicadas.")

if duplicados_antes > 0:
    df_limpio.drop_duplicates(inplace=True)
    logging.info("Filas duplicadas eliminadas.")
    print(f"Se eliminaron {duplicados_antes} filas duplicadas. Nuevo shape: {df_limpio.shape}")
else:
    logging.info("No se encontraron filas duplicadas para eliminar.")
```

Celda 5: Imputación de rango_estancia y Consolidación de Columnas


```python
# --- Celda 5 (Modificada): Imputar 'rango_estancia' usando 'tiempo_estancia_dias' ---

logging.info("Iniciando la imputación de 'rango_estancia' y la consolidación de columnas de estancia.")

# a) Manejo de outliers en 'tiempo_estancia_dias' antes de la conversión
# Acotamos los valores extremos a un máximo de 365 días para evitar categorías incorrectas.
df_limpio['tiempo_estancia_dias'] = df_limpio['tiempo_estancia_dias'].apply(
    lambda x: 365 if pd.notnull(x) and x > 365 else x
)
logging.info("Outliers en 'tiempo_estancia_dias' acotados a un máximo de 365.")

# b) Definir la función para mapear días a categorías de 'rango_estancia'
# Esta función toma un número de días y devuelve la etiqueta correspondiente.
def mapear_dias_a_rango(dias):
    if pd.isna(dias):
        return None
    if dias == 1:
        return '1 Noche'
    elif 2 <= dias <= 3:
        return '2-3 Noches'
    elif 4 <= dias <= 6: # Ajustado para que coincida con la lógica de los datos
        return '4-6 Noches'
    elif 7 <= dias <= 10:
        return '7-10 Noches'
    elif dias > 10:
        return 'Más de 10 noches'
    else:
        return None # Para casos como 0 días, que no deberían existir

# c) Aplicar la función para imputar los valores nulos en 'rango_estancia'
# Usamos .isnull() para identificar dónde 'rango_estancia' está vacío y 'tiempo_estancia_dias' tiene un valor.
condicion_imputar = df_limpio['rango_estancia'].isnull() & df_limpio['tiempo_estancia_dias'].notnull()

# Aplicamos la función solo a las filas que cumplen la condición
df_limpio.loc[condicion_imputar, 'rango_estancia'] = df_limpio.loc[condicion_imputar, 'tiempo_estancia_dias'].apply(mapear_dias_a_rango)

logging.info(f"Se imputaron {condicion_imputar.sum()} valores en 'rango_estancia' usando 'tiempo_estancia_dias'.")

# d) Convertir 'rango_estancia' a tipo categórico ordenado (ahora que está más completo)
orden_estancia = ['1 Noche', '2-3 Noches', '4-6 Noches', '7-10 Noches', 'Más de 10 noches', 'Reside']
# Aseguramos que los valores sean consistentes con las categorías definidas
df_limpio['rango_estancia'] = pd.Categorical(
    df_limpio['rango_estancia'], 
    categories=orden_estancia, 
    ordered=True
)
logging.info("La columna 'rango_estancia' ha sido convertida a tipo categórico ordenado.")

# e) Eliminar la columna numérica que ya no es necesaria
df_limpio.drop(columns=['tiempo_estancia_dias'], inplace=True)
logging.info("La columna 'tiempo_estancia_dias' ha sido eliminada.")

# --- Verificación ---
print("Verificación de la imputación en 'rango_estancia':")
print(df_limpio['rango_estancia'].value_counts(dropna=False)) # dropna=False para ver los nulos restantes

print("\nVerificación de las columnas del DataFrame tras la eliminación:")
print(list(df_limpio.columns))
```

Celda 6: Investigación del Problema en la Columna mes


```python
# --- 6. Investigar Anomalía en la Columna 'mes' ---
logging.info("Investigando los 14 valores únicos en la columna 'mes'.")

# Mostramos todos los valores únicos de la columna 'mes'
valores_unicos_mes = df_limpio['mes'].unique()
print(f"Valores únicos encontrados en 'mes':\n{valores_unicos_mes}")

# Generalmente, este problema es causado por:
# 1. Errores de tipeo (ej. 'Diciemrbe').
# 2. Espacios en blanco adicionales (' Diciembre ' vs 'Diciembre').
# 3. Diferentes formas de escribir el mismo mes (ej. 'Ene' vs 'Enero').

# Aplicamos una limpieza estándar: convertir a minúsculas y quitar espacios
df_limpio['mes'] = df_limpio['mes'].str.lower().str.strip()

# Volvemos a verificar después de la limpieza
valores_unicos_mes_limpio = df_limpio['mes'].unique()
print(f"\nValores únicos después de la limpieza básica:\n{valores_unicos_mes_limpio}")
print(f"\nNuevo conteo de valores únicos: {df_limpio['mes'].nunique()}")

# Si el problema persiste, podemos hacer mapeos específicos
# Ejemplo de mapeo si encontramos algo como 'diciem bre':
# df_limpio['mes'] = df_limpio['mes'].replace({'diciem bre': 'diciembre'})
```

```python
# --- Celda de Ingeniería de Características (CORRECCIÓN FINAL Y REUTILIZABLE) ---

logging.info("Creando nuevas características (features) para el modelo.")

# --- 1. Crear 'es_temporada_alta' basado en el mes ---
logging.info("Creando característica 'es_temporada_alta'...")

# SOLUCIÓN: Hacemos el código idempotente.
# Verificamos si 'desconocido' ya es una categoría antes de intentar añadirla.
if 'desconocido' not in df_limpio['mes'].cat.categories:
    df_limpio['mes'] = df_limpio['mes'].cat.add_categories(['desconocido'])
    logging.info("Categoría 'desconocido' añadida a 'mes'.")

df_limpio['es_temporada_alta'] = df_limpio['mes'].fillna('desconocido').apply(
    lambda x: 1 if x in ['diciembre', 'junio', 'julio'] else 0
).astype('Int8')

logging.info("Característica 'es_temporada_alta' creada exitosamente.")


# --- 2. Crear 'perfil_viajero' combinando características clave ---
logging.info("Creando característica 'perfil_viajero'...")

# Hacemos lo mismo para 'tipo_procedencia' y 'rango_edad'
if 'desconocido' not in df_limpio['tipo_procedencia'].cat.categories:
    df_limpio['tipo_procedencia'] = df_limpio['tipo_procedencia'].cat.add_categories(['desconocido'])
    logging.info("Categoría 'desconocido' añadida a 'tipo_procedencia'.")

if 'desconocido' not in df_limpio['rango_edad'].cat.categories:
    df_limpio['rango_edad'] = df_limpio['rango_edad'].cat.add_categories(['desconocido'])
    logging.info("Categoría 'desconocido' añadida a 'rango_edad'.")

tipo_proc_str = df_limpio['tipo_procedencia'].fillna('desconocido')
rango_edad_str = df_limpio['rango_edad'].fillna('desconocido')

df_limpio['perfil_viajero'] = tipo_proc_str.astype(str) + "_" + rango_edad_str.astype(str)
df_limpio['perfil_viajero'] = df_limpio['perfil_viajero'].astype('category')

logging.info("Característica 'perfil_viajero' creada exitosamente.")


# --- 3. Simplificar la estancia en categorías más manejables ---
logging.info("Creando característica 'tipo_estancia'...")

mapeo_estancia_simple = {
    '1 Noche': 'Estancia Corta',
    '2-3 Noches': 'Estancia Corta',
    '4-6 Noches': 'Estancia Media',
    '7-10 Noches': 'Estancia Media',
    'Más de 10 noches': 'Estancia Larga',
    'Reside': 'Residente'
}
df_limpio['tipo_estancia'] = df_limpio['rango_estancia'].map(mapeo_estancia_simple).astype('category')

logging.info("Característica 'tipo_estancia' creada exitosamente.")


# --- Verificación Final ---
logging.info("\nNuevas características creadas: 'es_temporada_alta', 'perfil_viajero', 'tipo_estancia'.")
print("Vista previa de las nuevas columnas:")
print(df_limpio[['mes', 'es_temporada_alta', 'perfil_viajero', 'rango_estancia', 'tipo_estancia']].head())

print("\nVerificando los tipos de datos de las nuevas columnas:")
print(df_limpio[['es_temporada_alta', 'perfil_viajero', 'tipo_estancia']].info())
```
